```python
##Funcion de ploteo
import matplotlib.pyplot as plt
def create_voltage_plot(data, xlabel, ylabel, title, figure_number=None):
    # Separate the data into x and y lists
    x_values, y_values = zip(*data)

    # Create a new figure if a figure number is specified
    if figure_number is not None:
        plt.figure(figure_number)
    
    # Create a line plot
    plt.plot(x_values, y_values)

    # Add labels and title
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)

    # Show the plot
    plt.grid()
    plt.show()
```
